When writing code, storing everything in one giant file creates problems. If you only want to test one small piece of code, you'd have to run the entire file from start to finish. Think of it like a recipe where you can test individual steps without making the whole dish. Between these code cells, you can add explanatory text (like this section) to guide yourself and others through your thinking.

This is especially powerful for data work, where you often need to try different approaches and see results immediately.

## 1.2: Python Basics


### Variables: Storing Information

Variables are containers that hold pieces of information. Think of them as labeled boxes where you store different types of data. Here are the main types you'll encounter:

- **Integer (int)**: Whole numbers like 10 or 100
- **Float**: Decimal numbers like 3.14
- **String**: Text like "Python" (always in quotes)
- **List**: A collection of values like [1, 2, 3]
- **Dictionary**: A collection where each piece of data has a label (key), like {"name": "Alice", "age": 20}

You can display what's stored in a variable by using the `print()` function, which shows the value on your screen.


```python
x = 10
y = 3.14
name = "Python"
my_list = [1, 2, 3]
my_dict = {"key": "value"}

print(x)
print(name)
```

### The Flexibility of Python

One of Python's superpowers is that variables are flexible. Once you create a variable, you can change what type of data it holds. In other languages, you'd have to declare a variable's type upfront. Python doesn't care about this. You could start with a number in a variable, then swap it out for text later. This gives you freedom but also requires careful attention, since you need to remember what each variable contains.


```python
print(type(name))
name = [1, 2, 3]
print(type(name))
name = {"key": "value"}
print(type(name))
```

Notice how we changed what the variable "name" contains three times? Each time, it updated to the new value. This is one of Jupyter's superpowers. The notebook remembers the current state of all your variables, so you can reference them later without redefining them.

### Loops: Repeating Actions

Loops let you perform the same action multiple times without writing the same code over and over. Imagine if you needed to process 100 items. Would you write 100 lines of code? No! You'd use a loop instead. You can loop through ranges of numbers, lists, or even individual characters in text.


```python
for num in range(5):
    print(f"Number: {num}")

name = "Python"
for letter in name:
    print(f"Letter: {letter}")
```

### Functions: Reusable Code Blocks

Functions are like shortcuts for code you use repeatedly. Instead of writing the same instructions five times, you write them once inside a function and call that function five times. This saves time, reduces errors, and makes your code cleaner.


```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))
print(greet("Bob"))
print(greet("Tom"))
```

### Error Handling: When Things Go Wrong

Unlike some languages that check your code before running it, Python runs your code line by line. This means errors might not appear until Python actually tries to execute a problematic line. If something goes wrong, the execution stops at that point and shows an error message. In the example below, we'll try to divide by zero, which will cause an error.


```python
print(greet("Alice"))

error = 10 / 0

print(greet("Bob"))
```

## 1.3: Introduction to NumPy ðŸ”¢


NumPy is a library built for numerical computing. At its core, it provides a powerful tool for storing and working with collections of data called **arrays**. Think of an array as an organized list of numbers that NumPy can work with very quickly, even if the list contains millions of items. This speed is crucial when processing large datasets.


```python
arr = np.array([1, 2, 3, 4, 5])
print("Array:", arr)
```

With arrays, you can do all sorts of powerful things:

- Select specific parts of the array (like picking certain items from a list)
- Perform math operations on every number at once (like multiplying all values by 2)
- Calculate summaries about the array (like finding the average or total)


```python
print("First element:", arr[0])
print("Slice:", arr[1:4])
```

```python
arr2 = arr * 2
print("Doubled Array:", arr2)
```

```python
print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))
```

NumPy also lets you create arrays without typing out every single number. You can generate arrays filled with zeros, ones, or numbers spaced evenly in a range.


```python
np.zeros(5)
```

```python
np.ones(5)
```

```python
np.linspace(0, 1, 5)
```

## 1.4: Working with Data Using Pandas ðŸ“Š


Pandas builds on NumPy's power and applies it to real world data. Think of it as Excel for Python. If you've ever worked with a spreadsheet, you understand Pandas. It stores data in tables with rows and columns, where each column can have a name (like "Age" or "Sales").

Pandas is perfect for loading data from files (like CSV files from Excel) and preparing it for analysis. You can quickly view your data, filter it, transform it, and get summaries, all with just a few lines of code.


```python
df = pd.read_csv('./DPA 01 - Intro Data.csv')

print(df.head())
```

Once you load data, you can do powerful things like selecting specific columns to work with or filtering the data to show only rows that match certain conditions.


```python
print(df[['X1']])
```

```python
print(df[['X2']])
```

```python
filtered_df = df[df['X1'] > 5]
print(filtered_df)
```
